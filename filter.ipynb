```python
import os

def process_md_files(folder_path, A_list, B):
    # 確認指定資料夾存在
    if not os.path.isdir(folder_path):
        print("資料夾不存在，請檢查路徑。")
        return
    
    # 遍歷資料夾中的所有 .md 檔案
    for filename in os.listdir(folder_path):
        if filename.endswith('.md'):
            file_path = os.path.join(folder_path, filename)
            
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()  # 讀取整個檔案為單一字串
            
            # 移除所有 A_list 中的多行字串
            for a in A_list:
                content = content.replace(a, "")
            
            # 檢查是否包含 B 並截斷內容
            if B in content:
                content = content.split(B)[0]
                found_B = True
            else:
                found_B = False
            
            # 將修改後的內容寫回檔案
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(content)
                
            # 輸出檔名和是否有 B
            if found_B:
                print(f"{filename} 有 {B}")
            else:
                print(f"{filename} 無 {B}")

# 使用範例
# folder_path = './problems'  # 指定資料夾路徑
folder_path = './problems/kamacoder/'  # 指定資料夾路徑
                
A_list = [
    '''<p align="center">
<a href="https://www.programmercarl.com/xunlian/xunlianying.html" target="_blank">
  <img src="../pics/训练营.png" width="1000"/>
</a>
<p align="center"><strong><a href="./qita/join.md">参与本项目</a>，贡献其他语言版本的代码，拥抱开源，让更多学习算法的小伙伴们受益！</strong></p>''', 
    '''<p align="center">
<a href="https://programmercarl.com/other/kstar.html" target="_blank">
  <img src="../pics/网站星球宣传海报.jpg" width="1000"/>
</a>'''
]  # 指定多行 A 字串
B = '## 其他语言版本'  # 指定變數 B
process_md_files(folder_path, A_list, B)
```
