```python
import requests
import pdfkit

def md_to_pdf(github_url, output_pdf):
    # 確保提供的 URL 是正確的 GitHub 原始內容鏈接
    if github_url.endswith('.md'):
        # 將 GitHub URL 轉換為原始內容 URL
        raw_url = github_url.replace('github.com', 'raw.githubusercontent.com').replace('/blob/', '/')
        response = requests.get(raw_url)
        
        if response.status_code == 200:
            # 獲取 Markdown 內容
            markdown_content = response.text
            
            # 將 Markdown 轉換為 HTML
            html_content = f"<html><body><pre>{markdown_content}</pre></body></html>"
            
            # 將 HTML 內容轉換為 PDF
            pdfkit.from_string(html_content, output_pdf)
            print(f"PDF 已成功保存為 {output_pdf}")
        else:
            print(f"無法獲取文件：{response.status_code}")
    else:
        print("請提供有效的 GitHub Markdown 文件 URL")

# 使用範例
github_markdown_url = 'https://github.com/adobe-fonts/source-han-serif/blob/release/README.md'  # 替換為實際的 URL
output_pdf_file = 'output.pdf'  # 替換為想要保存的 PDF 名稱
md_to_pdf(github_markdown_url, output_pdf_file)
```

```python
import markdown
import pdfkit

def md_file_to_pdf(md_file_path, output_pdf):
    try:
        # 讀取本地 Markdown 文件
        with open(md_file_path, 'r', encoding='utf-8') as md_file:
            markdown_content = md_file.read()
        
        # 將 Markdown 轉換為 HTML
        html_content = markdown.markdown(markdown_content)
        
        # 包裝成 HTML 文件結構
        full_html = f"<html><head><meta charset='utf-8'></head><body>{html_content}</body></html>"
        
        # 將 HTML 內容轉換為 PDF
        pdfkit.from_string(full_html, output_pdf)
        print(f"PDF 已成功保存為 {output_pdf}")
    except Exception as e:
        print(f"轉換失敗：{e}")

# 使用範例
local_md_file = './problems/0001.两数之和.md'  # 替換為本地的 Markdown 文件路徑
output_pdf_file = 'output.pdf'  # 替換為想要保存的 PDF 名稱
md_file_to_pdf(local_md_file, output_pdf_file)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pdfkit

# 設定 Selenium 驅動
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # 以無頭模式運行
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 目標 GitHub 頁面 URL
url = 'https://github.com/orange-36/leetcode-master-short/blob/main/problems/0203.%E7%A7%BB%E9%99%A4%E9%93%BE%E8%A1%A8%E5%85%83%E7%B4%A0.md'

# 打開頁面
driver.get(url)

# 等待頁面加載
driver.implicitly_wait(10)

# 抓取指定部分的內容（這裡假設你要抓取 README.md 的內容）
# content = driver.find_element(By.CSS_SELECTOR, 'div.markdown-body').get_attribute('innerHTML')
content = driver.find_element(By.XPATH, "/html/body/div[1]/div[5]/div/main/turbo-frame/div/react-app/div/div/div[1]/div/div/div[2]/div[2]/div/div[3]/div[2]")

# 保存內容為 HTML 文件
html_file_path = 'output.html'
with open(html_file_path, 'w', encoding='utf-8') as f:
    f.write(content)

# 使用 pdfkit 將 HTML 轉換為 PDF
pdf_file_path = 'output.pdf'
pdfkit.from_file(html_file_path, pdf_file_path)

# 關閉 Selenium 驅動
driver.quit()

print(f"PDF 已保存至 {pdf_file_path}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
from PIL import Image

# 設定 Selenium 驅動
options = Options()
options.add_argument('--headless')  # 無頭模式運行
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# 初始化 Chrome 驅動
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# 目標 URL
# 打開指定的網頁
driver.get(url)

# 等待頁面加載完成
time.sleep(3)

# 獲取頁面高度
last_height = driver.execute_script("return document.body.scrollHeight")
images = []

while True:
    # 截圖並保存
    screenshot_path = f'screenshot_{len(images)}.png'
    driver.save_screenshot(screenshot_path)
    images.append(screenshot_path)

    # 滾動到底部
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 等待加載
    time.sleep(2)

    # 獲取新高度並檢查是否已到達頁面底部
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 關閉驅動
driver.quit()

# 合併截圖
image_list = [Image.open(img) for img in images]
total_height = sum(img.size[1] for img in image_list)
max_width = max(img.size[0] for img in image_list)

# 創建一個新的空白圖片
merged_image = Image.new('RGB', (max_width, total_height))

# 將每個截圖添加到合併圖片中
y_offset = 0
for img in image_list:
    merged_image.paste(img, (0, y_offset))
    y_offset += img.size[1]

# 保存合併的圖片
merged_image.save('long_screenshot.png')

print("長截圖已保存至 long_screenshot.png")
```
